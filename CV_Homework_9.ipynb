```python
#--------------------------------------Runing the necessary imports--------------------------------------
#openCV
import cv2
# a library for generating random numbers
import random
# machine learning algorithms and tools
import dlib
#arrays and matrices
import numpy as np 
#visualizations
from matplotlib import pyplot as plt
#Change Plot Size
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
#--------------------------------------Loading an image.--------------------------------------
img = cv2.imread('data/emotions.jpg')
# grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#visualization
plt.imshow(gray, cmap='gray')
```

```python
#-------------------------------------- loading the detector--------------------------------------
detector = dlib.get_frontal_face_detector()
```

```python
#--------------------------------------Running the predictor on image.--------------------------------------
#face detection in the image
rects = detector(gray, 1)

def rect_to_bb(rect):
    # Dlib rect --> OpenCV rect
    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y

    return (x, y, w, h)
```

```python
#--------------------Drawing frames around detected faces and creating an image. Using different colors for each face.--------------------
result_dlib = np.copy(img)
faces_dlib_img = []
for rect in rects:    
    # Drawing rectangle around the face
    x, y, w, h = rect_to_bb(rect)
    color = random.sample(range(256), 3)
    cv2.rectangle(result_dlib, (x, y), (x+w, y+h), color, 2)
    faces_dlib_img.append(img[y:y+h, x:x+w, :])
# convert to RGB
img_rgb = cv2.cvtColor(result_dlib, cv2.COLOR_BGR2RGB)
#visualization
plt.imshow(img_rgb), plt.title('dlib')
```
