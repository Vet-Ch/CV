```python
import os
import cv2
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
# Load the training labels
root = 'data/'
data = pd.read_csv(os.path.join(root, 'Train.csv'))

# Number of training samples (amount of samples in data)
num_samples = len(data)

# Show random data samples
for ii in range(15):
    # Get random index
    idx = np.random.randint(0, num_samples)
    # Load image
    img = cv2.imread(os.path.join(root, data.iloc[idx]['Path']))
    # Convert image to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # Show image
    plt.subplot(3,5,ii+1), plt.imshow(img), plt.title(data.iloc[idx]['ClassId'])
```

```python
# Extract class identifiers
# Hint: Check the csv 
ids = data['ClassId']
```

```python
from collections import Counter
hist = Counter(ids)
plt.bar(hist.keys(), hist.values()), plt.grid(True)
plt.xlabel('Traffic Sign ID'), plt.ylabel('Counts')
```

```python
# According to the histogram, the data set is not balanced because some classes have much more samples than others
#There is a very large difference in the number of samples between some classes. Some have more than 2,000 samples,
#while others have less than 500
```
