```python
import os
import cv2
```

```python
#--------------------------------------------CSRT_tracker--------------------------------------------
vid = cv2.VideoCapture('data/video.mp4')

if not vid.isOpened():
    print("Video isn`t opened")
    exit()
    
# first frame
ret, frame = vid.read()
if not ret:
    print("frame isn`t readed")
    exit()
    
bbox = cv2.selectROI("Select object", frame)
cv2.destroyWindow("Select object")

# initializing CSRT
tracker = cv2.TrackerCSRT_create()
tracker.init(frame, bbox)

output_folder = "tracked_by_CSRT"
os.makedirs(output_folder, exist_ok=True)

fid = 0

while True:
    ret, frame = vid.read()
    if not ret:
        break
    
    ok, bbox = tracker.update(frame)
    
    if ok:
        x1, y1, w, h = [int(i) for i in bbox]
        cv2.rectangle(frame, (x1, y1), (x1 + w, y1 + h), (0, 255, 0), 2)
    else:
        print(f"not tracking on frame {fid}")
        
    cv2.imwrite(os.path.join(output_folder, f"frame_{fid:03d}.jpg"), frame)
    
    cv2.imshow("Tracked Frame", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    fid += 1
    
    if fid >= 15:
        break
    
vid.release()
cv2.destroyAllWindows()
```

```python
#--------------------------------------------MIL_tracker--------------------------------------------
vid = cv2.VideoCapture('data/video.mp4')

if not vid.isOpened():
    print("Video isn`t opened")
    exit()

# first frame
ret, frame = vid.read()
if not ret:
    print("frame isn`t readed")
    exit()

bbox = cv2.selectROI("Select object", frame)
cv2.destroyWindow("Select object")

# initializing MIL
tracker = cv2.TrackerMIL_create()
tracker.init(frame, bbox)

output_folder = "tracked_by_MIL"
os.makedirs(output_folder, exist_ok=True)

fid = 0

while True:
    ret, frame = vid.read()
    if not ret:
        break

    ok, bbox = tracker.update(frame)

    if ok:
        x1, y1, w, h = [int(i) for i in bbox]
        cv2.rectangle(frame, (x1, y1), (x1 + w, y1 + h), (0, 255, 0), 2)
    else:
        print(f"not tracking on frame {fid}")

    cv2.imwrite(os.path.join(output_folder, f"frame_{fid:03d}.jpg"), frame)

    cv2.imshow("Tracked Frame", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    fid += 1

    if fid >= 15:
        break

vid.release()
cv2.destroyAllWindows()
```

```python
#--------------------------------------------MIL_with_CSRT_tracker--------------------------------------------
vid = cv2.VideoCapture('data/video.mp4')

if not vid.isOpened():
    print("Video isn`t opened")
    exit()

# first frame
ret, frame = vid.read()
if not ret:
    print("frame isn`t readed")
    exit()

bbox1 = cv2.selectROI("Select object for CSRT", frame)
cv2.destroyWindow("Select object for CSRT")
bbox2 = cv2.selectROI("Select object for MIL", frame)
cv2.destroyWindow("Select object for MIL")

# initializing trakers
tracker_csrt = cv2.TrackerCSRT_create()
tracker_mil = cv2.TrackerMIL_create()

tracker_csrt.init(frame, bbox)
tracker_mil.init(frame, bbox)

output_folder = "tracked_by_CSRT_MIL"
os.makedirs(output_folder, exist_ok=True)

fid = 0

while True:
    ret, frame = vid.read()
    if not ret:
        break

    ok_csrt, bbox = tracker_csrt.update(frame)
    ok_mil, bbox = tracker_mil.update(frame)

    if ok_csrt:
        x1, y1, w1, h1 = [int(i) for i in bbox1]
        cv2.rectangle(frame, (x1, y1), (x1 + w1, y1 + h1), (0, 0, 255), 4)
    else:
        print(f"not tracking by CSRT on frame {fid}")
        
    if ok_mil:
        x2, y2, w2, h2 = [int(j) for j in bbox2]
        cv2.rectangle(frame, (x2, y2), (x2 + w2, y2 + h2), (0, 255, 0), 1)
    else:
        print(f"not tracking by MIL on frame {fid}")

    cv2.imwrite(os.path.join(output_folder, f"frame_{fid:03d}.jpg"), frame)

    cv2.imshow("Tracked Frame", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    fid += 1

    if fid >= 15:
        break

vid.release()
cv2.destroyAllWindows()
```

```python
#In this video, with detection at 15 frames, there is almost no difference. Both trackers stopped tracking the object when the camera approached it.
#When the object leaves the field of view, MIL continues tracking for some reason but ok_csrt becomes false
```
